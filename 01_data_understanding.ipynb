```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='is_fraud', data=df)
plt.title('Distribution of Fraudulent vs Non-Fraudulent Transactions')
plt.xlabel('Is Fraud')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x='is_fraud',
    y='velocity_score',
    data=df
)

plt.title('Fraud vs Non-Fraud Velocity Score')

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x='is_fraud',
    y='spending_deviation_score',
    data=df
)

plt.title('Fraud vs Non-Fraud Spending Deviation')

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    x='is_fraud',
    y='geo_anomaly_score',
    data=df
)

plt.title('Fraud vs Non-Fraud Geo Anomaly Score')

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    df[['amount',
        'time_since_last_transaction',
        'spending_deviation_score',
        'velocity_score',
        'geo_anomaly_score']].corr(),
    annot=True
)

plt.title('Feature Correlation Heatmap')

plt.show()
```
